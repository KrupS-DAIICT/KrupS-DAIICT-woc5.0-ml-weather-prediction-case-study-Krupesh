Import libraries

```python
import pandas as pd
import seaborn as sb
from matplotlib import pyplot as plt
```

Make dataframe of the CSV file.


```python
data = pd.read_csv('/content/drive/MyDrive/WoC\'23/Checkpoint-4/CarPrice_Assignment.csv')
data
```

<----- Graph 1 ---->

```python
# Plotting graph of enginetype vs horsepower using bars
plt.figure(figsize=(13, 8))
plt.style.use("fivethirtyeight")
width = 0.4
graph1 = plt.bar(data.enginetype, data.horsepower, width=width)
plt.title('Max Horsepower of given engines')
plt.xlabel('Engine type')
plt.ylabel('Horsepower(in hp) ')
```

<----- Graph 2 ----->

```python
plt.figure(figsize=(13, 8))
plt.style.use("fivethirtyeight")
width = 0.4
graph2 = plt.scatter(data.carlength, data.carwidth)
plt.xlabel('Length (in cm)')
plt.ylabel('Width (in cm)')
```

<----- Graph 3 ----->

```python
# Plotting Horsepower VS Car-price graph using seaborn KDEplot
plt.figure(figsize=(13, 8))
graph3 = sb.kdeplot(x='horsepower', y='price', data = data, shade=True, cbar=True)
graph3.set(xlabel="Horsepower (in hp)", ylabel="Car price (in USD)")
```

<----- Graph 4 ----->

```python
# Plotting the graph of Cylinder number VS Horsepower with seaborn Voiline plot graph
plt.figure(figsize=(13, 8))
data = data[data.cylindernumber.isin(['four', 'six', 'five', 'eight'])]
data = data[data.drivewheel.isin(['rwd', 'fwd'])]
graph4 = sb.violinplot(x = 'cylindernumber', y = 'horsepower', hue='drivewheel',
                       data = data, linewidth=1.5)
graph4.set(xlabel="Cylinders", ylabel="Horsepower (in hp)")
```

<----- Graph 5 ----->

```python
plt.figure(figsize=(13, 8))
graph5 = sb.stripplot(x = 'peakrpm', y = 'fueltype',
                      hue='cylindernumber', data = data, 
                      jitter=0.15, size=6, linewidth=1)
```
